```python
import tensorflow as tf
import numpy as np
```

```python
resistencia=np.array([2,3,4,5,6,7,8,9], dtype=float)
invcorriente=np.array([1.11,1.67,2,2.5,3.3,3.3,5,5], dtype=float)
```

```python
import matplotlib.pyplot as plt
plt.xlabel("Resistencia ($Ohm$)")
plt.ylabel("Inverso de la corriente ($Amp^{-1}$)")
#plt.plot(resistencia,invcorriente)

plt.plot(resistencia,0.5783*resistencia-0.1958)
plt.plot(resistencia,invcorriente,**{'color': 'lightsteelblue', 'marker': 'o'})
plt.show()
```

```python
capa = tf.keras.layers.Dense(units=1, input_shape=[1])
modelo = tf.keras.Sequential([capa])

# oculta1 = tf.keras.layers.Dense(units=3, input_shape=[1])
# oculta2 = tf.keras.layers.Dense(units=3)
# salida = tf.keras.layers.Dense(units=1)
# modelo = tf.keras.Sequential([oculta1, oculta2, salida])
```

```python
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss='mean_squared_error'
)
```

```python
print("Comenzando entrenamiento...")
historial = modelo.fit(resistencia,invcorriente, epochs=1000, verbose=False)
print("Modelo entrenado!")
```

```python
plt.xlabel("# Epoca")
plt.ylabel("Magnitud de pérdida")
plt.plot(historial.history["loss"])
```

```python
print("Hagamos una predicción!")
resultado = modelo.predict([9])
print("El resultado es " + str(resultado) + "$Amp^{-1}$!")
```
